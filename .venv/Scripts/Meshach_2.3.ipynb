## Creating a Line plot of the Tempertures for 2022

```python
df_temp = df_citi.set_index('DATE')
```

```python
df_temp.head()
```

```python
# Sort the DataFrame by the DATE column in ascending order
df_temp.sort_values(by='DATE', ascending=True, inplace=True)
```

```python
# Use the Pandas built-in plot function directly
plot = df_temp['TAVG'].plot(
    title='Average Temperature for 2022',
    figsize=(12, 6),  # Adjust the size to better fit the data
    color='blue',     # You can choose a color that fits your style
)
plot.set_xlabel('Date')  # Setting the x-label
plot.set_ylabel('Average Temperature (C)')  # Setting the y-label
plot.grid(True)  # Adding a grid for better readability
```

## A column with the count of trips per day

```python
df_temp.head()
```

```python
# Groupby the count of the trips 

df_group = pd.DataFrame(df_temp.groupby(['DATE'])['ride_id'].count()).reset_index()
```

```python
df_group.head()
```

```python
df_group.rename(columns = {'ride_id':'bike_trips_daily'}, inplace = True)
df = df_citi.merge(df_group, on = "DATE", how = 'outer', indicator = True)
print(df['_merge'].value_counts(dropna = False))
print("Shape of January 1st is", df[df['DATE'] == '2022-01-01'].shape) # Check 
print("Shape of January 2nd is", df[df['DATE'] == '2022-01-02'].shape) # Second check 

df_temp = df.set_index('DATE')

print(df_temp.columns)
```

```python
df_temp.head()
```

```python
# Set up the figure with a specific size
plt.figure(figsize=(10, 5))
# Add a title to the plot
plt.title('Temperature per ride')
# Plot the average temperature from the dataframe
plt.plot(df_temp['TAVG'], label='Average Temperature')  # Plot the 'TAVG' data as a line plot
# Plot daily bike trips in red color from the dataframe
plt.plot(df_temp['bike_trips_daily'], color='red', label='Daily Bike Trips')  # Plot 'bike_trips_daily' data in red
# Display the plot
plt.show()
```

```python
# Create figure and axis objects with subplots()
fig,ax = plt.subplots(figsize=(10, 5))
# Make a plot of the bike rides
ax.plot(df_temp['bike_trips_daily'],color = "red")
# Here is where we create the dual axis - initialize a twin object for two different y-axis on the sample plot
ax2 = ax.twinx()
# Make a plot with the temperatues on the y-axis using second axis object
ax2.plot(df_temp["TAVG"], color = "blue")
plt.show()
```

## Dual Plot

```python
# Create figure and axis objects with subplots() and details
fig,ax = plt.subplots(figsize=(10, 5))

# Make a plot of the bike rides
ax.plot(df_temp['bike_trips_daily'],color = "navy")
ax.set_xlabel("Year 2022", fontsize = 14) # set x-axis label
ax.set_ylabel("Bike trips daily",color = "navy", fontsize = 14) # set y-axis label

# Here is where we create the dual axis - initialize a twin object for two different y-axis on the sample plot
ax2 = ax.twinx()

# Make a plot with the temperatues on the y-axis using second axis object
ax2.plot(df_temp["TAVG"], color = "red")
ax2.set_ylabel("Average temperatures", color = "red",fontsize=14)
plt.title('Temperature and trips in 2022', fontsize = 18)
plt.show()
```

```python
# Create figure and axis objects with subplots() and details
fig,ax = plt.subplots(2, figsize =(8, 6))

# Make a plot of the bike rides
ax[0].plot(df_temp['bike_trips_daily'],color = "navy")
ax[0].set_ylabel("Bike trips daily",color = "navy", fontsize = 14) # set y-axis label

# Make a plot with the temperatues on the y-axis using second axis object
ax[1].plot(df_temp["TAVG"], color = "red")
ax[1].set_ylabel("Average temperatures", color = "red",fontsize=14)
plt.suptitle('Temperature and trips in 2022', fontsize = 18)

plt.show()
```

### Explanation of the Dual-Axis Line Plot

In this visualization, I used Matplotlib, a powerful library for creating static, interactive, and animated visualizations in Python. Matplotlib provides a flexible way to create complex plots with detailed customizations. Heres how the plot was constructed:

#### 1. **Creating the Figure and Axes**
- The `subplots()` function was used to initiate a figure and a single subplot. This function returns a `figure` and `axes` object. The `figsize` parameter was set to ensure the plot is wide enough to comfortably display a full year of data.

#### 2. **Plotting the Primary Data**
- The primary data series (daily bike trips) was plotted on the primary y-axis using the `plot()` method on the `axes` object (`ax`). This data was set in a navy color to visually differentiate from the temperature data.

#### 3. **Setting Axis Labels and Titles**
- Labels for both the x-axis and the primary y-axis were defined with appropriate font sizes for clarity. The x-axis represents the year 2022, and the primary y-axis represents the count of daily bike trips.

#### 4. **Creating a Secondary Axis**
- A secondary y-axis was created using the `twinx()` method on the original `axes` object. This method creates a new axes (`ax2`) that shares the same x-axis but has a separate y-axis, making it ideal for plotting data with different scales.

#### 5. **Plotting the Secondary Data**
- The secondary data series (average temperatures) was plotted on this secondary y-axis in red, providing a clear visual contrast to the primary data series.

#### 6. **Enhancing the Plot with a Title**
- A title was added to the plot to provide context, using a larger font size for emphasis and clarity.

#### 7. **Displaying the Plot**
- Finally, `plt.show()` was called to display the plot. This method renders the plot to the screen, allowing for visual inspection of the combined data trends.

### Paradigm Used
The Object-Oriented Paradigm of Matplotlib was utilized here. This approach is more flexible and recommended for complex visualizations because it allows direct control over the figure and axes objects. It facilitates customization and manipulation of multiple layers and axes, which is crucial for detailed, multifaceted visualizations such as this dual-axis plot.
