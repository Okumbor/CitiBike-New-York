## Dual-axis line plot

```python
df['DATE'] = pd.to_datetime(df['DATE'])
```

```python
df_temp = df.set_index('DATE')
```

```python
# Sort the DataFrame by the DATE column in ascending order
df_temp.sort_values(by='DATE', ascending=True, inplace=True)
```

```python
# Groupby the count of the trips 
df_group = pd.DataFrame(df_temp.groupby(['DATE'])['ride_id'].count()).reset_index()
```

```python
df_group.rename(columns = {'ride_id':'bike_trips_daily'}, inplace = True)
df = df.merge(df_group, on = "DATE", how = 'outer', indicator = True)
print(df['_merge'].value_counts(dropna = False))
print("Shape of January 1st is", df[df['DATE'] == '2022-01-01'].shape) # Check 
print("Shape of January 2nd is", df[df['DATE'] == '2022-01-02'].shape) # Second check 

df_temp = df.set_index('DATE')

print(df_temp.columns)
```

```python
# Set the aesthetic style of the plots to 'ticks'
sns.set(style="ticks")  
# Create figure and axis objects with subplots()
fig, ax = plt.subplots(figsize=(10, 6))
# Make a plot of the bike rides using Seaborn's styling
sns.lineplot(data=df_temp, x=df_temp.index, y='bike_trips_daily', color="red", ax=ax)
ax.set_xlabel('Year 2022')  # Optional: Set x-axis label if needed
ax.set_ylabel('Bike trips daily', color="red")  # Set the y-axis label and color
# Create a dual axis using Matplotlib's twinx for the temperature
ax2 = ax.twinx()
# Make a plot with the temperatures on the y-axis using the second axis object
sns.lineplot(data=df_temp, x=df_temp.index, y="TAVG", color="blue", ax=ax2)
ax2.set_ylabel('Average Temperature (Â°C)', color="blue")  # Set the y-axis label and color
# Optionally, you can improve the layout and/or aesthetics further
fig.tight_layout()  # Adjust layout to make room for labels or tick labels
# Setting the super title for the figure
plt.title('Temperature and trips in 2022', fontsize=12, pad=20)
# Display the plot
plt.show()
```

```python
# Set the aesthetic style of the plots to 'ticks'
sns.set(style="ticks") 
# Create figure and axis objects with subplots()
fig, ax = plt.subplots(2, 1, figsize=(10, 6))  # Adjust subplot to 2 rows, 1 column
# Make a plot of the bike rides
sns.lineplot(data=df_temp, x=df_temp.index, y='bike_trips_daily', color="navy", ax=ax[0])
ax[0].set_ylabel("Bike trips daily", color="navy", fontsize=14)  # set y-axis label
# Adjust the spacing between the subplots
plt.subplots_adjust(hspace=0.5)  # Adjusts the horizontal space between plots
# Make a plot with the temperatures on the y-axis using second axis object
sns.lineplot(data=df_temp, x=df_temp.index, y="TAVG", color="red", ax=ax[1])
ax[1].set_ylabel("Average temperatures", color="red", fontsize=14)
# Setting the super title for the figure
plt.suptitle('Temperature and trips in 2022', fontsize=18)
# Show the plot
plt.show()
```

```python
plt.figure(figsize=(9,5))
sns.boxplot(x = 'rideable_type', y = 'bike_trips_daily', data = df, palette = {'electric_bike': 'royalblue', 'classic_bike': 'pink', 'docked_bike': 'green'})
plt.show()
```

## Detailed Analysis of Box Plot Elements

### Central Tendency and Spread
- Each box represents the interquartile range (IQR) for daily trips, indicating the middle 50% of the data for each bike type.
- The line within each box denotes the median number of trips.
- The electric and classic bikes show a higher and more compact IQR compared to docked bikes, suggesting more consistent usage.

### Outliers
- The plot for docked bikes shows several outliers, depicted as individual points below the lower whisker.
- These outliers indicate days with significantly fewer trips than typical for docked bikes, suggesting sporadic usage or operational issues on those days.

### Whiskers
- The whiskers extend from the boxes to the highest and lowest values that are not considered outliers.
- The wide range of the whiskers, especially for the docked bikes, indicates high variability in daily trips for this type, possibly reflecting fluctuating demand or availability issues.

### Comparison Among Types
- Comparatively, electric and classic bikes have a higher median and more consistent usage than docked bikes.
- This could imply a preference or higher availability of these types in daily operations.


# FacetGrid

```python
grid = sns.FacetGrid(df, col = "rideable_type")
grid.map(sns.histplot, "bike_trips_daily", bins = 15)
plt.show()
```

## Histogram Analysis of Daily Bike Trips by Rideable Type

- **Electric Bikes**: The distribution of daily trips for electric bikes is somewhat uniform but slightly skewed towards lower trip counts, suggesting a steady but varied use across different days.
- **Classic Bikes**: Classic bikes show a clear peak around 2000 to 3000 trips per day, indicating a more consistent and possibly preferred usage pattern compared to other types.
- **Docked Bikes**: The histogram for docked bikes shows a very sparse distribution, with most of the counts close to zero. This suggests that docked bikes are significantly less utilized compared to electric and classic bikes.
